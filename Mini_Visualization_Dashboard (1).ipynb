Merging the two datasets

```python
d1=pd.read_csv("student-mat.csv",sep=";")
d2=pd.read_csv("student-por.csv",sep=";")

d3 = pd.merge(d1, d2, on=[
    "school","sex","age","address","famsize","Pstatus",
    "Medu","Fedu","Mjob","Fjob","reason","nursery","internet"
])
```

#Features

1) Average Score of both subject

```python
d3['Avg_score']=(d3['math_finale_score']+d3['portuguese_finale_score'])/2
```

2) Performance Category

```python
# Maximum score is 20
def perfo_cat(score):
  if score>15:
    return 'High'
  elif score>8:
    return "Medium"
  else:
    return "Low"

d3['performance_category']=d3["Avg_score"].apply(perfo_cat)
```

3) Score Difference

```python
d3['score_diff']=d3['math_finale_score']-d3['portuguese_finale_score']
```

4) Studytime

```python
#average studytime
d3['avg_stdtime']=(d3['studytime_x']+d3['studytime_y'])/2
```

```python
d3['studytime_category']=d3['avg_stdtime'].map({1:'<2 hours',2:'2-5 hours',3:'5-10 hours',4:'>10 hours'})
```

5) Absence level

```python
def level(x):
  if x<=5:
    return 'Low'
  elif x<=15:
    return "Medium"
  else:
    return "High"

d3['absence_level']=d3['absences_x'].apply(level)
```

#**Visualization**

Factors Affecting Student Performance

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.makedirs("images", exist_ok=True)
```

##**1) Score Distribution**

   - Plot: Histogram of Avg_score
   - Purpose: Understand overall performance spread

```python
plt.figure()
plt.hist(data=d3,x='Avg_score',bins=20,edgecolor='black')
plt.xlabel("Average Score")
plt.xticks(np.arange(0,20,2))
plt.ylabel("Count")
plt.title("Score Distribution")
plt.savefig("images/avg_score_distribution.png",bbox_inches="tight")
plt.show()
```

**Interpretation:**

Most students have average scores in the mid range, with fewer students achieving very high or very low scores. This indicates a moderate overall performance level.

##**2) Study Time vs Performance**

   - Plot: Bar plot
(studytime_category vs Avg_score)
   - Purpose: Does study time affect performance?

```python
plt.figure()
# Calculate the mean Avg_score for each studytime_category
grouped_data = d3.groupby('studytime_category')['Avg_score'].mean().reset_index()
# Sort the categories for better visualization
category_order = ['<2 hours', '2-5 hours', '5-10 hours', '>10 hours']
grouped_data['studytime_category'] = pd.Categorical(grouped_data['studytime_category'], categories=category_order, ordered=True)
grouped_data = grouped_data.sort_values('studytime_category')

plt.bar(x=grouped_data['studytime_category'], height=grouped_data['Avg_score'])
plt.xlabel("Study Time")
plt.ylabel("Average Score")
plt.title("Study Time vs Performance")
plt.grid(True,axis="y")
plt.savefig("images/study_time_vs_performance.png",bbox_inches="tight")
plt.show()
```

**Interpretation:**

Students who study more hours tend to have higher average scores, indicating a positive relationship between study time and academic performance.

##**3) Absences vs Performance**

   - Plot: Boxplot
(absence_level vs Avg_score)
   - Purpose: Do absences impact performance?

```python
plt.figure()
sns.boxplot(d3,x='absence_level',y='Avg_score')
plt.savefig("images/Absence_Vs_Performance.png")
plt.show()
```

**Interpretation:**

Students with higher absences show lower median scores, suggesting that regular attendance plays a key role in better academic outcomes.

##**4) Math vs Portuguese**

   - Plot: Scatter plot
(math_finale_score vs portuguese_finale_score)
   - Purpose: Are performances in subjects related?

```python
plt.figure()
sns.scatterplot(d3,x='math_finale_score',y='portuguese_finale_score',color='orange')
plt.xlabel("Maths Score")
plt.ylabel("Portuguese Score")
plt.title("Math vs Portuguese")
plt.savefig("images/Maths_vs_portuguese_scatter.png")
plt.show()
```

**Interpretation**

There is a strong positive relationship between math and Portuguese scores, indicating that students performing well in one subject tend to perform well in the other.

##**5) Performance Category Distribution**

   - Plot: Countplot
(performance_category)
   - Purpose: How many students are high/medium/low performers?

```python
plt.figure()
sns.countplot(d3,x='performance_category',)
plt.xlabel('Performance')
plt.title("Performance Category Distribution")
plt.savefig("images/Performance_Category_Distribution.png")
plt.show()
```

**Interpretation:**

Most students fall into the medium performance category, while fewer students are classified as high or low performers.

##**6) Failures vs Score**

   - Plot: Boxplot
(failures_x vs Avg_score)
   - Purpose: Do past failures affect performance?

```python
plt.figure()
sns.boxplot(d3,x='failures_x',y='Avg_score',color='g')
plt.xlabel('number of past class failures')
plt.ylabel('Averge Score')
plt.title('Failures vs Score')
plt.savefig("images/Failures_vs_Score.png")
plt.show()
```

**Interpretetion:**

Students with more past failures tend to have lower average scores, indicating that academic history strongly influences future performance.

##**7) Correlation Heatmap**

   - Plot: Heatmap
   - Purpose: To understand relationships between numeric features

```python
corr = d3[[
    "math_finale_score",
    "portuguese_finale_score",
    "Avg_score",
    "studytime_x",
    "failures_x",
    "absences_x",
    "Dalc_x",
    "Walc_x"
]].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig("images/correlation_heatmap.png")
plt.show()
```

**Interpretation:**

There is a strong positive correlation between math and Portuguese scores, indicating that students performing well in one subject tend to perform well in others. Failures show a negative correlation with scores, suggesting that past academic struggles impact current performance.

Absences and alcohol consumption show weak to moderate negative relationships with performance, indicating lifestyle factors may affect academic outcomes.

#Conclusion
